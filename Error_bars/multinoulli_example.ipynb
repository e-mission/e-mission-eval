```python
car = 1.512; eb = 0.022; b = 0
EI = [car,eb,b]
P = [0.7,0.2,0.1]
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from matplotlib.ticker import MultipleLocator
```

```python
#figure(figsize = (8,6), dpi=80)

fig, ax = plt.subplots(1, figsize=(8,8))

ax.plot(EI,P,'bo')
#plt.xticks([0,0.022,1.512],["0: walk/bike","ebike", "car"])
#plt.gca().xaxis.set_major_locator(MultipleLocator(0.25))

#[0,0.022, 1.512], ["walk/bike", "ebike", "car"]

ax.vlines([0,0.022, 1.512],ymin= [0,0,0], ymax=[0.1,0.2,0.7], colors = "black")
#ax.xticks(list(ax.xticks()[0]) + list(plt.xticks([1.512],["car"])[0]))
ax.set_ylabel("P(EI | predicted car)")
ax.set_xlabel("Energy Intensity (EI) in kWH/PMT")
ax.set_xbound([-0.05, 1.8])
ax.set_title("Probability mass function for energy intensity values given that we predicted car")

fig.subplots_adjust(bottom=0.5)
```
